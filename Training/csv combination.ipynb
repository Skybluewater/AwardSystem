```python
def to_csv_head(csv_name, csv_head, class_num=3):
    df_record = pd.read_csv(csv_name, header=0, encoding='utf-8')
    csv_head_n = df_record.columns[1: -1]
    if "label_encoding" in csv_name:
        csv_name_s = csv_name.split("_label")[0]
    else:
        csv_name_s = csv_name.split("_fact")[0]
    csv_head.append(csv_name_s + "_count")
    for i in range(0, class_num):
        for k in csv_head_n:
            csv_head.append(csv_name_s + "_" + k + "_" + str(i))
    return csv_head

csv_head = ['user_id', 'award_id']
csv_head = to_csv_head("paper_label_encoding_spectral_clustering.csv", csv_head)
csv_head = to_csv_head("patent_record_fact.csv", csv_head)
csv_head = to_csv_head("research_award_label_encoding_spectral_sorted.csv", csv_head)
csv_head = to_csv_head("research_project_fact.csv", csv_head)
csv_head = to_csv_head("competition_award_label_encoding_spectral.csv", csv_head)
csv_head = to_csv_head("honor_title_fact.csv", csv_head)
csv_head = to_csv_head("other_achievements_fact.csv", csv_head)
```

```python
df_all = pd.DataFrame(columns=csv_head)
df_all['user_id'] = df_user['user_id']
df_all['award_id'] = df_user['awardName']

def filling_label(df_all, df_single_name):
    df_single = pd.read_csv(df_single_name, header=0, encoding='utf-8').drop(["Unnamed: 0"], axis=1)
    count_num = -1
    index = 0
    previous_user_id = -1
    if "label_encoding" in df_single_name:
        name_s = df_single_name.split("_label")[0]
    else:
        name_s = df_single_name.split("_fact")[0]
    for ix, row in df_single.iterrows():
        
        # 很扯淡,但是确实如此,有的人奖学金数据为空,没有去掉这些人的论文等信息,所以手动处理一下,手动删除那几行就是王道,3889 & 3897, 并且防止长度过长进行修正
        if row['user_id'] == 3897 or row['user_id'] == 3889:
            continue
        
        if row['user_id'] == previous_user_id: # previous_user_id = df_all.loc[index, ['user_id']]
            count_num += 1
        else:
            list_name = name_s + "_count" 
            df_all.loc[index, list_name] = count_num + 1
            count_num = 0
        
        while df_all.loc[index, :]['user_id'] != row['user_id']:
            index += 1
#             print("a" + str(df_all.loc[index, :]['user_id']))
        
        def fillin():
            column_names = df_single.columns[: -1]
            for column in column_names:
                column_name = name_s + "_" + column + "_" + str(count_num)
                df_all.loc[index, column_name] = row[column]
                if "label" in column:
                    df_all.loc[index, column_name] += 1
        if count_num <= 2:
            fillin()
        previous_user_id = row['user_id']
    list_name = name_s + "_count" 
    df_all.loc[index, list_name] = count_num + 1
    return df_all.copy()
```

```python
df_all = filling_label(df_all, "paper_label_encoding_spectral_clustering.csv")
df_all = filling_label(df_all, "patent_record_fact.csv")
df_all = filling_label(df_all, "competition_award_label_encoding_clustering_sorted.csv")
df_all = filling_label(df_all, "honor_title_label_encoding_sorted.csv")
df_all = filling_label(df_all, "research_award_label_encoding_spectral_sorted.csv")
df_all = filling_label(df_all, "research_project_label_encoding_sorted.csv")
df_all = filling_label(df_all, "other_achievements_label_encoding_sorted.csv")
```
